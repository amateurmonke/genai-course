Importing transformers library

```python
from transformers import pipeline, set_seed, GPT2Tokenizer
```

```python
set_seed(42)
```

```python
prompt = "The future of Artificial Intelligence is"
```

# BERT: Experiment 1

```python
bert = pipeline('text-generation', model='distilbert-base-uncased')

# Generate text
output_bert = bert(prompt, max_length=50, num_return_sequences=1)
print(output_bert[0]['generated_text'])
```

BERT is an encoder-only model, and thus not designed for text generation. However, it is possible to generate text with it by adding a mask. BERT is capable of predicting the next word in a sentence.

OBSERVATION: The next word predicted for the sequence is "uncertain", giving us the sentence "The future of Artificial Intelligence is uncertain."

# RoBERTa: Experiment 1

```python
roberta_generator = pipeline('text-generation', model='roberta-base')

output_roberta = roberta_generator(prompt, max_length=50, num_return_sequences=1)
print(output_roberta[0]['generated_text'])
```

RoBERTa is also an encoder-only model, thus even it is not capable of generating text.

# BART: Experiment 1

```python
bart = pipeline('text-generation', model='facebook/bart-base')

# Generate text
output_bart = bart(prompt, max_length=50, num_return_sequences=1)
print(output_bart[0]['generated_text'])
```

OBSERVATION: The model generated gibberish.

# BERT: Experiment 2

```python
prompt_exp2 = "The goal of Generative AI is to [MASK] new content"
```

```python
bert = pipeline('fill-mask', model='distilbert-base-uncased')

# Generate text
output_bert = bert(prompt_exp2, top_k = 1)
print(output_bert[0]['token_str'])
```

OBSERVATION: The word predicted is "generate", giving us the sentence "The goal of Generative AI is to generate new content."

# RoBERTa: Experiment 2

```python
# roberta has a different mask token
prompt_exp2_roberta = "The goal of Generative AI is to <mask> new content"
```

```python
roberta = pipeline('fill-mask', model='roberta-base')

output_roberta = roberta(prompt_exp2_roberta, top_k = 1)
print(output_roberta[0]['token_str'])
```

OBSERVATION: The generated text is "The goal of Generative AI is to create new content."

# BART: Experiment 2

```python
bart = pipeline('fill-mask', model='facebook/bart-base')

output_bart = bart(prompt_exp2_roberta, top_k = 1)
print(output_bart[0]['token_str'])
```

OBSERVATION: The predicted word is "create", giving us the sentence "The goal of Generative AI is to create new content."

# BERT: Experiment 3

```python
prompt_qa = "Generative AI poses significant risks such as hallucinations, bias and deepfakes."
q = "What are the risks?"
```

```python
bert = pipeline('question-answering', model = 'distilbert-base-uncased')
output_bert = (bert(question = q, context = prompt_qa))
print(output_bert['answer'])
```

```python
# squad

bert2 = pipeline('question-answering', model = 'distilbert-base-cased-distilled-squad')
output_bert2 = (bert2(question = q, context = prompt_qa))
print(output_bert2['answer'])
```

both versions of BERT give us the same result: The risks of GenAI are hallucinations, bias and deepfakes.

# RoBERTa: Experiment 3

```python
roberta_qa = pipeline('question-answering', model=('roberta-base'))
output_roberta_qa = (roberta_qa(question = q, context = prompt_qa))
print(output_roberta_qa['answer'])
```

```python
roberta_qa2 = pipeline('question-answering', model=('deepset/roberta-base-squad2'))
output_roberta_qa2 = (roberta_qa2(question = q, context = prompt_qa))
print(output_roberta_qa2['answer'])
```

# BART: Experiment 3

```python
bart_qa = pipeline('question-answering', model='facebook/bart-base')

# Generate text
output_bart_qa = (bart_qa(question = q, context = prompt_qa))
print(output_bart_qa['answer'])
```

```python
bart_qa2 = pipeline('question-answering', model='a-ware/bart-squadv2')

# Generate text
output_bart_qa2 = (bart_qa2(question = q, context = prompt_qa))
print(output_bart_qa2['answer'])
```
