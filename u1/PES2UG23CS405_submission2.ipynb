```python
from transformers import pipeline
import pdfplumber
```

```python
qa = pipeline(
    "question-answering",
    model="distilbert-base-cased-distilled-squad"
)
```

```python
def read_pdf(path):
    text = ""
    with pdfplumber.open(path) as pdf:
        for page in pdf.pages:
            page_text = page.extract_text()
            if page_text:
                text += page_text + "\n"
    return text
```

```python
context = read_pdf("sample_genai.pdf")
print(context[:500])
```

```python
question = "What enabled the boom of generative AI in recent years?"
qa(question=question, context=context)
```

```python
question = "What are some applications of generative AI?"
qa(question=question, context=context)
```

```python
question = "Whem are gemerative AI systems multimodal?"
qa(question=question, context=context)
```
